```python
from IPython.display import clear_output
import time

# Game variables
board = [[1,2,3],[4,5,6],[7,8,9]]
current_player_char = 'X'
next_player_char = 'O'
totalInputs = 9
winner = None

def checkHorizontal():
    # Check each row for same char
    for row in board:
        if row[0] == row[1] == row[2] and row[0] in ['X','O']:
            return True
    return False

def checkVertical():
    # Check each column
    for c in range(3):
        if board[0][c] == board[1][c] == board[2][c] and board[0][c] in ['X','O']:
            return True
    return False

def checkDiagonal():
    # Two diagonals
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] in ['X','O']:
        return True
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] in ['X','O']:
        return True
    return False

def checkBoard():
    # If any win condition is True
    return checkHorizontal() or checkVertical() or checkDiagonal()

def placeCharacterOnBoard(pos):
    global current_player_char

    # Validate position
    if pos < 1 or pos > 9:
        print("Invalid position! Choose 1â€“9.")
        time.sleep(1)
        return 0

    # Convert position to row/col
    row = (pos - 1) // 3
    col = (pos - 1) % 3

    # Check if already taken
    if board[row][col] in ['X','O']:
        print("Position already taken! Try again.")
        time.sleep(1)
        return 0

    # Place character
    board[row][col] = current_player_char
    return 1

def printCurrentBoard():
    print("-------------")
    for eachRow in board:
        print("|", end="")
        for eachCol in eachRow:
            print(f" {eachCol} ", end="|")
        print()
        print("-------------")

def runGame():
    global current_player_char
    global next_player_char
    global winner

    print("Welcome to TIC-TAC-TOE!")

    inputCount = 0

    while inputCount < totalInputs:
        printCurrentBoard()

        position = int(input(f"Player {(inputCount % 2) + 1}, enter a free position: "))

        validity = placeCharacterOnBoard(position)
        inputCount += validity

        if inputCount >= 5:
            if checkBoard():
                winner = "Player 1" if current_player_char == 'X' else "Player 2"
                clear_output()
                break

        if validity:
            current_player_char, next_player_char = next_player_char, current_player_char

        clear_output()

    printCurrentBoard()
    if winner:
        print(f"Well done, {winner}. You have won the game!")
    else:
        print("The game ended in a draw.")

runGame()
```
